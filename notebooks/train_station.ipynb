```python
# get all the geocode by google api
import googlemaps
api_key='Your api key'
gmaps = googlemaps.Client(api_key)

import pandas as pd

df = pd.read_csv('../data/curated/Curated_Rental_Property_Data.csv')

print(df)
col= ['distance to melb central in km']
dis= pd.DataFrame(columns=col)
print(dis)
a=[]
cord=[]
loc2=[144.961602, -37.813937]
cord.append(loc2)



print(df['address'])

for i in df['address']:
    result = gmaps.geocode(i)
    lat= result[0]["geometry"]['location']['lat']
    lang=result[0]["geometry"]['location']['lng']
    loc1=[lang,lat]
    
    
    cord.append(loc1)


    #a.append(hs.haversine(loc1,loc2))
dis['distance to melb central in km']=a
print(cord)
```
